```python
import gexp
```

```python
gexp.download_data(cancer_list=['LUAD','LUSC','ACC'], data_source='./cancer_link.csv', data_dir ='Data')
```

```python
ori_data = gexp.load_labeled_data(data_dir = './Data', label_list=['LUAD','LUSC','ACC'])
ori_data
```

```python
data = ori_data.copy()

#data.iloc[:,:-1] = gexp.normalize(data.iloc[:,:-1], methods = ['log1p'])

data['Target'] = data['Target'].map({'LUAD':0, 'LUSC':1, 'ACC':2})
```

```python
data
```

```python
rank, importances = gexp.biomarker_rank(data, models=[['RF', 'default'],['XGB' , {'colsample_bytree': 0.5, 'n_estimators': 200, 'subsample': 0.75}],['Ada','recommended']])
```

```python
rank
```

```python
importances
```

```python
import numpy as np
acc = gexp.plot_stepwise_accuracy(data, rank, list(np.arange(1,10)), model = ['RF', "recommended"], accuracy_metric=['accuracy'], multi_class=True)
acc
```

```python
log1p_zscore_df = gexp.normalize(ori_data, methods = ['log1p','z_score'], exclude='Target')

gene_list = rank.sort_values(by='RF').iloc[:10].index
gexp.plot_heatmap(log1p_zscore_df, gene_list)
```
