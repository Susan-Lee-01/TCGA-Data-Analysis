```python
#binary classification
```

```python
gexp.download_data(cancer_list=['LUAD','LUSC'], data_source='./cancer_link.csv', data_dir ='Data')
```

```python
binary_data = gexp.load_labeled_data(data_dir = './Data', label_list=['LUAD', 'LUSC'])
binary_data
```

```python
data = binary_data.copy()

#data.iloc[:,:-1] = gexp.normalize(data.iloc[:,:-1], methods = ['log1p'])

data['Target'] = data['Target'].map({'LUAD':0, 'LUSC':1})
```

```python
data
```

```python
rank, importances = gexp.biomarker_rank(data, models=[['RF', 'default'],['XGB' , {'colsample_bytree': 0.5, 'n_estimators': 200, 'subsample': 0.75}],['Ada','recommended']])
```

```python
rank
```

```python
importances
```

```python
import numpy as np
acc = gexp.plot_stepwise_accuracy(data, rank, list(np.arange(1,10)), model = ['RF', "recommended"], accuracy_metric=['accuracy'])
acc
```

```python
log1p_zscore_df = gexp.normalize(binary_data, methods = ['log1p','z_score'], exclude='Target')

gene_list = rank.sort_values(by='RF').iloc[:10].index
gexp.plot_heatmap(log1p_zscore_df, gene_list)
```

```python
#multi classification
```

```python
gexp.download_data(cancer_list=['BRCA'], data_source='./cancer_link.csv', data_dir ='Data')
```

```python
# 유방암 옵션은 ['LumA, 'Her2', 'LumB', 'Basal'] 
multi_data = gexp.load_labeled_data(data_dir = './Data', label_list=['BRCA_LUMA','BRCA_LUMB','BRCA_HER2','BRCA_BASAL'], patient_type = './BRCApatients_type.csv')
multi_data
```

```python
data = multi_data.copy()

#data.iloc[:,:-1] = gexp.normalize(data.iloc[:,:-1], methods = ['log1p'])

data['Target'] = data['Target'].map({'BRCA_LUMA':0, 'BRCA_LUMB':1, 'BRCA_HER2':2, 'BRCA_BASAL':3})
```

```python
data
```

```python
rank, importances = gexp.biomarker_rank(data, models=[['RF', 'default'],['XGB' , {'colsample_bytree': 0.5, 'n_estimators': 200, 'subsample': 0.75}],['Ada','recommended']])
```

```python
rank
```

```python
importances
```

```python
import numpy as np
acc = gexp.plot_stepwise_accuracy(data, rank, list(np.arange(1,10)), model = ['MLP', "recommended"], accuracy_metric=['accuracy'], multi_class=True)
acc
```

```python
log1p_zscore_df = gexp.normalize(multi_data, methods = ['log1p','z_score'], exclude='Target')

gene_list = rank.sort_values(by='RF').iloc[:10].index
gexp.plot_heatmap(log1p_zscore_df, gene_list)
```
