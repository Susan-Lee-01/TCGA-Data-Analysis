```python
import gexp
```

```python
Double_df = gexp.load_labeled_data(data_dir = './Data', label_list=['LUAD', 'LUSC'], patient_type = './BRCApatient_type.csv')
Double_df
```

```python
# 유방암 옵션은 ['LumA, 'Her2', 'LumB', 'Basal'] 
Multi_df = gexp.load_labeled_data(data_dir = './Data', label_list=['BRCA_LUMA','BRCA_LUMB','BRCA_HER2','BRCA_BASAL'], patient_type = './BRCApatient_type.csv')
Multi_df
```

```python
# feature 노말라이즈
log1p_df = gexp.normalize(Multi_df, methods = ['log1p'], exclude = 'Target')
log1p_df
```

```python
# feature 노말라이즈
z_score_df = gexp.normalize(Multi_df, methods = ['z_score'], exclude = 'Target')
z_score_df
```

```python
# feature 노말라이즈
log1p_z_score_df = gexp.normalize(Multi_df, methods = ['log1p', 'z_score'], exclude = 'Target')
log1p_z_score_df
```

```python
# 타겟 문자->숫자 인코딩
from sklearn.preprocessing import LabelEncoder
df_log1p = log1p_df.copy()
df_log1p['Target'] = LabelEncoder().fit_transform(df_log1p['Target'])
df_log1p
```

```python
# 타겟 문자->숫자 인코딩
from sklearn.preprocessing import LabelEncoder
df_z_score = z_score_df.copy()
df_z_score['Target'] = LabelEncoder().fit_transform(df_z_score['Target'])
df_z_score
```

```python
# 타겟 문자->숫자 인코딩
from sklearn.preprocessing import LabelEncoder
df_log1p_z_score = log1p_z_score_df.copy()
df_log1p_z_score['Target'] = LabelEncoder().fit_transform(df_log1p_z_score['Target'])
df_log1p_z_score
```

```python
rank_log1p, importance_log1p = gexp.biomarker_rank(df_log1p, models = [['RF', 'default'],['XGB' , {'colsample_bytree': 0.5, 'n_estimators': 200, 'subsample': 0.75}]])
```

```python
rank_log1p
```

```python
importance_log1p
```

```python
rank_z_score, importance_z_score = gexp.biomarker_rank(df_z_score, models = [['RF', 'default'],['XGB' , {'colsample_bytree': 0.5, 'n_estimators': 200, 'subsample': 0.75}]])
```

```python
rank_z_score
```

```python
importance_z_score
```
